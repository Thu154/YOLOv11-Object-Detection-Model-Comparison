```python
from ultralytics import YOLO
from IPython.display import Image, display
import matplotlib.pyplot as plt
import cv2
import os
```

```python
source_image = '/content/buses.jpeg'
model_path1 = 'yolo11s.pt'
model_path2 = 'yolo11m.pt'
model_path3 = 'yolo11l.pt'
model_path4 = 'yolo11x.pt'

# Load and predict using the first model
model1 = YOLO(model_path1)
results1 = model1.predict(source=source_image, conf=0.5, save=True)

# Load and predict using the second model
model2 = YOLO(model_path2)
results2 = model2.predict(source=source_image, conf=0.5, save=True)

# Load and predict using the second model
model3 = YOLO(model_path3)
results3 = model3.predict(source=source_image, conf=0.5, save=True)

# Load and predict using the second model
model4 = YOLO(model_path4)
results4 = model4.predict(source=source_image, conf=0.5, save=True)


```

```python
# Load images after prediction
result_image1 = cv2.imread('/content/runs/detect/predict/buses.jpg')
result_image2 = cv2.imread('/content/runs/detect/predict2/buses.jpg')
result_image3 = cv2.imread('/content/runs/detect/predict3/buses.jpg')
result_image4 = cv2.imread('/content/runs/detect/predict4/buses.jpg')

# Convert BGR to RGB for display
result_image1 = cv2.cvtColor(result_image1, cv2.COLOR_BGR2RGB)
result_image2 = cv2.cvtColor(result_image2, cv2.COLOR_BGR2RGB)
result_image3 = cv2.cvtColor(result_image2, cv2.COLOR_BGR2RGB)
result_image4 = cv2.cvtColor(result_image2, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(14, 10))

# List of results and titles
result_images = [result_image1, result_image2, result_image3, result_image4]
titles = ['YOLOv11-S', 'YOLOv11-M', 'YOLOv11-L', 'YOLOv11-X']

for i, image in enumerate(result_images):
    plt.subplot(2, 2, i + 1) # 2 rows, 2 columns
    plt.imshow(image)
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()

```

```python
# Gathering key metrics for 4 models
metrics = {
    'Model': ['YOLO11-S', 'YOLO11-M', 'YOLO11-L', 'YOLO11-X'],
    'Inference Time (ms)': [
        results1[0].speed['inference'],
        results2[0].speed['inference'],
        results3[0].speed['inference'],
        results4[0].speed['inference']
    ],
    'Objects Detected': [
        len(results1[0].boxes),
        len(results2[0].boxes),
        len(results3[0].boxes),
        len(results4[0].boxes)
    ],
    'Confidence Threshold': [0.5, 0.5, 0.5, 0.5]
}

# Convert to DataFrame
import pandas as pd
metrics_df = pd.DataFrame(metrics)

# Display results as a table
print(metrics_df)
```
